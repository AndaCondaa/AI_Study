# Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5, 5, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
# ReLU 
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5, 5, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c) # for overflow
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    
    return y
```
